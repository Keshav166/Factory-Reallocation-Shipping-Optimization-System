```python
products = df_original[['Product Name', 'Factory']].drop_duplicates()

print(products.shape)
```

```python
products.shape
```

```python
recommendations = []

for _, row in products.iterrows():

    product = row['Product Name']
    current_factory = row['Factory']

    for factory in all_factories:

        if factory != current_factory:

            recommendations.append([
                product,
                current_factory,
                factory
            ])

recommendations = pd.DataFrame(
    recommendations,
    columns=[
        'Product',
        'Current Factory',
        'Alternative Factory'
    ]
)

print(recommendations.head())
```

```python
print(recommendations.shape)
```

```python
recommendations["Score"] = (
    recommendations["Avg Profit"] * 10
)
```

```python
recommendations = recommendations.sort_values(
    by="Score",
    ascending=False
)

print(recommendations.head(10))
```

```python
product_profit = df_original.groupby(
    "Product Name"
)["Gross Profit"].mean()
```

```python
recommendations["Avg Profit"] = recommendations["Product"].map(
    product_profit
)
```

```python
final_recommendations = (
    recommendations
    .sort_values("Score", ascending=False)
    .groupby("Product")
    .first()
    .reset_index()
)

print(final_recommendations)
```
